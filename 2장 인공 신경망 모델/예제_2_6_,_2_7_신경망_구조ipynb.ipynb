예제 2.6, 2.7 신경망 구조

```python
# 예제 2.6 단층 신경망을 이용한 분류

# 셋업
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten

# 모델 생성
model = Sequential([
    Flatten(input_shape=(3, 3), name='Input_Layer'), # 2차원 패턴을 1차원 벡터로 변환
    Dense(2, name='Output_Layer')   # 출력층 뉴런 2개
])

# 모델 구조 요약
model.summary()
```

```python
# 예제 2.7 3계층 신경망을 이용한 분류

# 모델 생성(model.add() 메소드 이용)
model = Sequential()
model.add(Flatten(input_shape=(9, 7), name='Input_Layer'))
model.add(Dense(2, name='Hidden_Layer'))
model.add(Dense(3, name='Output_Layer'))

# 모델 구조 요약
model.summary()
```

```python
# 모델 구조
tf.keras.utils.plot_model(model)
```

```python
# 모델 구조
tf.keras.utils.plot_model(model, show_shapes=True)
```
