예제 2.4 McCulloch-Pitts 모델(AND)

```python
# 예제 2.4 McCulloch-Pitts 모델의 AND 연산

import numpy as np

# McCulloch-Pitts 모델의 AND 연산 정의
def and_gate(x):
    w = np.array([1, 1])   # 가중치를 학습하지 않고 1로 고정
    net = np.dot(x, w)

    if net >= 2:   # 임계치 2
       return 1
    else:
       return 0

# AND 연산 검증
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

for i in range(4):
    y = and_gate(x[i])
    print(f"x:{x[i]}  y:{y}")   # print("x:{} y:{}".format(x[i], y))와 동일
```
